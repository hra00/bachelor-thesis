```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
```

```python
ENV = ['9*9 w/o subgoals', '9*9 with subgoals', '11*11 w/o subgoals', '11*11 with subgoals', '13*13 w/o subgoals', '13*13 with subgoals']
```

```python
success_probability = np.array([1.0, 1.0, 0, 1.0, 0, 0.95])
time_step = np.array([30.1, 13.2, 100, 17.0, 100, 37.9])
discounted_return = np.array([0.753408193,0.875622030, 0, 0.842935747,0, 0.695705994])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(28,7))
success_probability_df = pd.DataFrame(data = success_probability, index=ENV, columns=[''])
success_probability_df.plot(kind="bar", title="success probability", ax=axes[0], alpha=0.5, rot=60)
time_step_df = pd.DataFrame(data = time_step, index=ENV, columns=[''])
time_step_df.plot(kind="bar", title="time step", ax=axes[1], alpha=0.5, rot=60)
discounted_return_df = pd.DataFrame(data = success_probability, index=ENV, columns=[''])
discounted_return_df.plot(kind="bar", title="discounted return", ax=axes[2], alpha=0.5, rot=60)
```

```python
success_probability = np.array([1.0, 1.0, 0, 1.0, 0, 0.95])
time_step = np.array([30.1, 13.2, 100, 17.0, 100, 37.9])
discounted_return = np.array([0.753408193,0.875622030, 0, 0.842935747,0, 0.695705994])

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(28,7))
success_probability_df = pd.DataFrame({ 'celltype':["9*9 Gridworld","11*11 Gridworld","13*13 Gridworld"], 'without subgoals':[1,0,0], 'with subgoals':[1,1,0.95]})
time_step_df = pd.DataFrame({ 'celltype':["9*9 Gridworld","11*11 Gridworld","13*13 Gridworld"], 'without subgoals':[30.1,100,100], 'with subgoals':[13.2, 17, 37.9]})
discounted_return_df = pd.DataFrame({ 'celltype':["9*9 Gridworld","11*11 Gridworld","13*13 Gridworld"], 'without subgoals':[0.75, 0, 0], 'with subgoals':[0.88, 0.84, 0.70]})

success_probability_df = success_probability_df[["celltype","without subgoals","with subgoals"]]
time_step_df = time_step_df[["celltype","without subgoals","with subgoals"]]
discounted_return_df = discounted_return_df[["celltype","without subgoals","with subgoals"]]

success_probability_df.set_index(["celltype"],inplace=True)
time_step_df.set_index(["celltype"],inplace=True)
discounted_return_df.set_index(["celltype"],inplace=True)

success_probability_df.plot(kind="bar", title="success probability", ax=axes[0], alpha=0.7, rot=0, xlabel="")
axes[0].legend(loc='lower left')
for container in axes[0].containers:
    axes[0].bar_label(container)
time_step_df.plot(kind="bar", title="time step", ax=axes[1], alpha=0.7, rot=0, xlabel="")
axes[1].legend(loc='upper left')
for container in axes[1].containers:
    axes[1].bar_label(container)
discounted_return_df.plot(kind="bar", title="discounted return", ax=axes[2], alpha=0.7, rot=0, xlabel="")
axes[2].legend(loc='upper right')
for container in axes[2].containers:
    axes[2].bar_label(container)
```
